### Note on hedonic regression & price modeling
- Hedonic regression is a commonly used to understand pricing thanks to its flexibility and applicability to economic questions. It can be easily transformed into a utility curve through derivation, and can be applied to willingness-to-pay at price indices.
- Because our central focus is simply to optimize our price model, we're going to look to hedonic regression to guide attribute selection and modification.
- That being said, let's review the central components of a basic pricing regression
    - Squaring Total Square Feet attribute (common practice in hedonic regression to "account for the non-linear relationship betwee size of the building and price" - Attribute Selection Guide)
    - Class: up to 3 classes can be developed with binary using selected characteristics.
    - Land size
    - Outdoor space
    - Parking type
    - Year Renovated (better than age of building according to Attribute Selection Guide)
    - Construction type as a class
    - Urban versus Suburban
    - log(yi) to reduce variance from the getgo and better determine restrictions for models

## OUTLIERS

```python
def count_those_outliers(dataset, **attributes):
    '''
    this function takes in a dataset and designated attributes
    and returns the number of outliers with a useful messaege.
    Dataset = an interable
    attributes = list of column names. Formatted as such ['age', 'time'].
    If excluded, function iterates through columns searching for numeric values.
    '''
    
    if type(attributes)==str:
        attributes = [attributes]
    else:
        attributes = dataset.select_dtypes(exclude = ['object'])
    #including this to eliminate any columns that contain non-numeric values
    
    #need to insert a pass for any non-numeric columns
    for column_name in attributes:
        n = len(dataset)
        outliers = len(dataset.loc[(dataset[column_name] >
                              (np.mean(dataset[column_name]) + (np.std(dataset[column_name])*3)))
                             | (dataset[column_name] <
                                (np.mean(dataset[column_name]) -( np.std(dataset[column_name])*3)))])
        ratio = outliers/n
        if outliers == 0:
            pass
        elif outliers > 0 and ratio < 0.1:
            print(column_name, 'has', outliers, 'outliers out of',
                  n, 'objects.\nYou can confidently eliminate these outliers.\n')

        elif outliers > 0 and ratio > 0.1:
            print('Uh oh!', column_name, 'has', outliers, 'out of',
                  n, 'objects!\nYou should consider an alternative null-value methodology.\n')
        else:
            print('Please check your inputs.\n')
```

In other words, our function did not return any flags. Additionally, these outliers will become less important once we develop a second or third model.

```python
#using this functionset from StackOverflow to autofill any classifications not in the datasets
def add_missing_dummy_columns(d, columns ):
    '''
    d = dataframe of dummy variables
    columns = all possible categories/classifications
    '''
    missing_cols = set( columns ) - set( d.columns )
    for c in missing_cols:
        d[c] = 0
```

```python
#secondary component to the function set
def fix_columns( d, columns ):  
    '''
    d = dataframe of dummy variables
    columns = list of possible classifications
    '''

    add_missing_dummy_columns( d, columns )

    # make sure we have all the columns we need
    assert( set( columns ) - set( d.columns ) == set())

    extra_cols = set( d.columns ) - set( columns )
    if extra_cols:
        print("extra columns:", extra_cols)

    d = d[ columns ]
    return d
```

```python
def get_corr(t_df,t_target):  
    '''
    This takes in your dataframe and generates correlations
    t_df = your full dataframe
    t_target = your yi, formatted df['Target']
    '''
    t_corr_list=list()  
    t_col_count=t_df.shape[1]  
    for i in range(0,t_col_count):  
       #print('in corr')  
        t_corr_list.append(t_df.iloc[:,i].corr(t_target))  
    return t_corr_list

#save output from this function for the next
```

```python
def get_features_filtered_on_corr(t_f, t_corr, t_thresh):  
    '''
    t_f = list of features
    t_corr = list of correlations from get_corr function
    t_thresh = threshold desired
    '''
    t_width=t_f.shape[1]  
    tB=pd.DataFrame(np.asarray(t_corr),columns=['corr'])  
    tB=tB.fillna(0)    
    t_bool=tB['corr']>t_thresh  
    t_range=list(range(0,t_width))  
    t_bool2=pd.concat([pd.Series(t_bool),pd.Series(t_range)],axis=1)  
    t_bool3=t_bool2[t_bool]  
    t_good=list(t_bool3.iloc[:,1])  
    tF_good=t_f.iloc[:,t_good]  
    return (tF_good)  

```
